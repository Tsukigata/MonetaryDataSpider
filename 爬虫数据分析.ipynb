```python
import os
import pandas as pd
```

```python
#所使用的数据是沪市A股的按股票代码排序的前80只股票在2021-10-22的日线行情，数据来源于对新浪财经的历史成交明细
#获取文件列表信息
base_filepath = 'D:\\jupyter_notebook\\Findata'
filenames = os.listdir(base_filepath)
filenames
```

```python
#读入文件
with open(base_filepath + '\\' + filenames[1],'r',encoding = "utf-8") as f:
    data = f.readlines()
    print(data)
f.close()
```

```python
#将文件转化为DataFrame
Data = []
for line in data:
    Data.append(line.strip('\n').split('\t'))
print(Data)
data = pd.DataFrame(Data,columns = ['code','Date','time','price','price_change','deal_volume','deal_amount','attribute','meaningless'])
data
```

```python
#去重
data.drop_duplicates()
```

```python
#数据排序
data.sort_values(by="time",ascending=True)
```

```python
#批量操作，将各个股票的文件存储到DataFrame构成的列表中
stocks = []
for filename in filenames:
    with open(base_filepath + '\\' + filename,'r',encoding = "utf-8") as f:
        data = f.readlines()
        #print(data)
    f.close()
    Data = []
    for line in data:
        Data.append(line.strip('\n').split('\t'))
    data = pd.DataFrame(Data,columns = ['code','Date','time','price','price_change','deal_volume','deal_amount','attribute','meaningless'])
    stocks.append(data)
    print(data)
```

```python
stocks[0]
```

```python
#数据去重排序
for stock in stocks:
    stock.drop_duplicates(inplace = True)
    stock.sort_values(by="time",ascending=True,inplace = True)
    print(stock.describe())
```

```python
#交易频率（每天总的交易次数）
deal_frequenlity = []
for stock in stocks:
    deal_frequenlity.append(stock.shape[0])
```

```python
#交易频率最大的公司
filenames[deal_frequenlity.index(max(deal_frequenlity))]
```

```python
#交易频率最小的公司
filenames[deal_frequenlity.index(min(deal_frequenlity))]
```

```python
#成交量总量（单位：手）
total_volume = []
for stock in stocks:
    total_volume.append(stock['deal_volume'].astype('int').sum())
```

```python
#最大的成交量总量的股票
filenames[total_volume.index(max(total_volume))]
```

```python
max(total_volume)
```

```python
#最小成交量总量的股票
filenames[total_volume.index(min(total_volume))]
```

```python
min(total_volume)
```

```python
#稳定性分析（以价格的标准差来衡量）
stability = []
for stock in stocks:
    stability.append(stock['price'].astype('float').std())
```

```python
#价格变动幅度最大的股票
filenames[stability.index(max(stability))]
```

```python
#价格变动幅最度小的股票
filenames[stability.index(min(stability))]
```
